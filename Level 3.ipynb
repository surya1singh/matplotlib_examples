# Basemap

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

m = Basemap(projection='mill')
m.drawcoastlines()
plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

m = Basemap(projection='mill',
            llcrnrlat = -40,
            llcrnrlon = -40,
            urcrnrlat = 50,
            urcrnrlon = 75)
m.drawcoastlines()
plt.show()
```

```python
m.drawcoastlines()
m.drawcountries(linewidth=2)
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

m = Basemap(projection='mill',
            llcrnrlat = -90,
            llcrnrlon = -180,
            urcrnrlat = 90,
            urcrnrlon = 180)

m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')

plt.title('Basemap Tutorial')
plt.show()
```

```python
m = Basemap(projection='mill',
            llcrnrlat = -10,
            llcrnrlon = 60,
            urcrnrlat = 40,
            urcrnrlon = 100,
            resolution='h')
#c - crude
#l - low
#h - high
#f - full

m.drawcoastlines()
m.drawcountries(linewidth=2)
#m.etopo()
m.bluemarble()

plt.title('Basemap Tutorial')
plt.show()
```
